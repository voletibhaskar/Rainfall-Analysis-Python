```python
plt.figure(figsize=(100,100))

for column_index, column in enumerate(rain_data.columns):
    if column == 'STATE_UT_NAME':
        continue
    plt.subplot(6,6, column_index + 1)
    sb.violinplot(x='STATE_UT_NAME', y=column, data=rain_data)
```

```python
plt.figure(figsize=(100, 100))

for column_index, column in enumerate(rain_data.columns):
    if column == 'STATE_UT_NAME':
        continue
    plt.subplot(6, 6, column_index + 1)
    sb.boxplot(x='STATE_UT_NAME', y=column, data=rain_data)
```
